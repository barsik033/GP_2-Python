```python
api["clean_address"] = (
    api["address"]
    .astype(str)
    .str.replace(r"(?i)россия,?\s*", "", regex=True)
    .str.replace(r"(?i)москва,?\s*", "", regex=True)
    .str.replace(r"(?i)мск,?\s*", "", regex=True)
    .str.replace(r"(?i)рф,?\s*", "", regex=True)
    .str.strip(", ")
)

api["street"] = api["clean_address"].apply(lambda x: ",".join(x.split(",")[:-1]).strip() if "," in x else x)
api["house_number"] = api["clean_address"].apply(lambda x: x.split(",")[-1].strip() if re.search(r'\d', x.split(",")[-1]) else "")

api.loc[~api["house_number"].str.contains(r"\d"), "house_number"] = ""
api.loc[api["house_number"] == "", "street"] = api["clean_address"]

```

```python
api['address'] = api['address'].astype(str)

corpus_patterns = [
    r',\s*к\b',
    r',\s*корп\.?\b',
    r',\s*кор\.?\b',
    r',\s*кп\.?\b'
]

ctr_patterns = [
    r',\s*с\b',
    r',\s*стр\.?\b',
    r',\s*ст\.?\b',
    r',\s*строение\.?\b'

]

def clean_address(x):
    x = x.lower().strip()

    for pattern in corpus_patterns:
        x = re.sub(pattern, ' к', x)

    for pattern in ctr_patterns:
        x = re.sub(pattern, ' с', x)

    x = re.sub(r'\s{2,}', ' ', x)
    return x.strip()

api["address"] = api["address"].apply(clean_address)
```

```python
api['House number'] = api['House number'].astype(str)

corpus_replacements = [r'корпус', r'корп\.?', r'кор\.?', r'кп\.?', r'к\.?']

building_replacements = [r'строение', r'стр\.?', r'ст\.?', r'с\.?']

house_prefixes = [r'д\.?', r'дом', r'номер', r'№']

def clean_house_number(x):
    x = str(x).lower().strip()
    x = re.sub(r'\s+', '', x)

    for pattern in house_prefixes:
        x = re.sub(pattern, '', x)

    for pattern in corpus_replacements:
        x = re.sub(pattern, 'к', x)

    for pattern in building_replacements:
        x = re.sub(pattern, 'с', x)

    x = re.sub(r'(?<=[a-zа-яё])\.', '', x, flags=re.IGNORECASE)

    x = re.sub(r'к+', 'к', x)
    x = re.sub(r'с+', 'с', x)

    x = re.sub(r'[^0-9а-яa-zкс]', '', x)

    return x

api['House number'] = api['House number'].apply(clean_house_number)
```
